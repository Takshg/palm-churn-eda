```python
import importlib.metadata
import os
import sys
import tempfile
import types
from pathlib import Path

import pandas as pd

os.environ.setdefault("MPLCONFIGDIR", tempfile.mkdtemp())

try:
    import pkg_resources
except ModuleNotFoundError:
    pkg_resources = types.ModuleType("pkg_resources")

    class _Distribution:
        def __init__(self, name):
            self.version = importlib.metadata.version(name)

    def get_distribution(name):
        return _Distribution(name)

    pkg_resources.get_distribution = get_distribution
    sys.modules["pkg_resources"] = pkg_resources

from ydata_profiling import ProfileReport

OUTPUT_DIR = Path("Dataset_Profile_Reports")
OUTPUT_DIR.mkdir(parents=True, exist_ok=True)

def report_generation(df, name):
    # Generate and save the profile report for the given dataset.
    profile = ProfileReport(df, title="Dataset Profiling Report", explorative=True)
    output_path = OUTPUT_DIR / f"{name}_profile_report.html"
    profile.to_file(output_path)
    return output_path

emp = pd.read_csv("data/BOLT_Employees.csv")
applicant = pd.read_csv("data/BOLT_Applicants.csv")
changes = pd.read_csv("data/BOLT_EmployeeChanges.csv")
perf = pd.read_csv("data/BOLT_Performance.csv")
branch = pd.read_csv("data/BOLT_Branch.csv")

report_generation(emp, "employee")
report_generation(applicant, "applicant")
report_generation(changes, "employee_changes")
report_generation(perf, "performance")
report_generation(branch, "branch")
```
