```python
import pandas as pd
from sqlalchemy import create_engine

df = pd.read_csv('employee_master.csv') 
engine = create_engine('sqlite://', echo=False)

df.to_sql('emp', con=engine, if_exists='replace', index=False)
```

```python
sql_query = 'SELECT COUNT(EmployeeID) FROM emp WHERE HighestEducationLevel = "PhD"' 
df_results = pd.read_sql_query(sql_query, con=engine)
print(df_results.head())
```

```python
sql_query = 'SELECT COUNT(EmployeeID) FROM emp WHERE HighestEducationLevel = "Bachelor"' 
df_results = pd.read_sql_query(sql_query, con=engine)
print(df_results.head())

sql_query = 'SELECT COUNT(EmployeeID) FROM emp WHERE HighestEducationLevel = "Bachelor" AND Churned = 1' 
df_results = pd.read_sql_query(sql_query, con=engine)
print(df_results.head())
```
